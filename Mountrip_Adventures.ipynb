```python
df["amount"] = df["amount"].fillna(df["num_people"] * df.groupby(["package","nights"])["amount"].transform(lambda x: (x / df.loc[x.index, "num_people"]).median()))
```

```python
#finding per person amount

df["amount_per_person"] = df["amount"] / df["num_people"]
```

```python
from sqlalchemy import create_engine
from sqlalchemy.engine import URL

url = URL.create(
    drivername="mysql+pymysql",
    username="root",
    password="ashok@1234", 
    host="localhost",
    port=3306,
    database="mountrip_adventures"
)

engine = create_engine(url)

print("✅ Connected successfully")
```

```python
# pushing data to mysql 

df.to_sql(
    name="raw_data",
    con=engine,
    if_exists="replace",  
    index=False
)

print("✅ Data successfully loaded into MySQL table: raw_data")
```

```python
#verifiying data

test_df = pd.read_sql("SELECT * FROM raw_data LIMIT 5;", engine)
print(test_df)
```

```python
#fixing the data types

df["travel_date"] = pd.to_datetime(df["travel_date"], dayfirst=True, errors="coerce")
df["amount"] = df["amount"].astype(int)
df["num_people"] = df["num_people"].astype(int)
df["nights"] = df["nights"].astype(int)
```
